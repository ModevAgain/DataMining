```python
popsize = 100
nofattributes = 42
used_attributes = 20
max_iterations = 50

### CREATION
# first population random
p = Population(42,20,popsize)
p.random_populate()

iter_best = []
iteration = 0

for i in range(0,max_iterations):
    print("iteration"+str(iteration))
    print("population mean fitness:"+str(p.getMeanFitness()))
    best = p.getMaxFitness()
    print("population max fitness individual: " +str(best.fitness))
    iter_best.append((iteration,best,best.fitness))
    iteration+=1
    p = p.getNextGeneration_multicore()
```

```python
from genetic_algorithm_music_clustering import *
popsize = [25,50,100,10]
used_attributes = list(range(10,26))
iterations = 30
nofcores = 4

experiments = []

nofexperiments = len(popsize)*len(used_attributes)
finished_experiments = 0

for pops in popsize:
    for ua in used_attributes:
        print("starting experiment with "+str(pops)+" individuals and "+str(ua)+" used attributes...")
        e = experiment(nofcores,pops,ua,iterations)
        e.start()
        experiments.append(e)
        finished_experiments += 1
        print("finished experiment "+str(finished_experiments)+" - "+str(nofexperiments-finished_experiments)+" remaining...")
```

```python
experiments[0].plot()
```

```python
attributes = experiments[0].alltime_best_fitness[-1][0].attribute_settings
```

```python
import pandas as pd
df1 = pd.DataFrame.from_csv("FeatureFileTrainingAllList1.csv")
df2 = pd.DataFrame.from_csv("FeatureFileTestAllList2.csv")

import sklearn.preprocessing as skp
import numpy as np

for df in [df1,df2]:
    for i in range(0,len(df.columns)):
        df.iloc[:,i] = skp.scale(df.iloc[:,i])

        
from itertools import compress
bol_list = []
for a in attributes:
    if a>0:
        bol_list.append(True)
    else:
        bol_list.append(False)

ua = list(compress(df1.columns.tolist(), bol_list))

import copy
dfc = copy.deepcopy(df1)

for col in dfc.columns.tolist():
    if col not in ua:
        del dfc[col]
        
dfc.to_csv("subFeaturesTrain1.csv")
```
